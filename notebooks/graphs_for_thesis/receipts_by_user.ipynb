```python
import numpy as np
import pandas as pd
import seaborn as sns

sns.set()

# every user submits one receipt a day
receipts_per_month = 30

users = [1000 * i for i in range(13)]

receipts = [0]
for i in range(1, 13):
    receipts.append(receipts[i-1] + users[i] * receipts_per_month)
    
data = pd.DataFrame({"users": users, "receipts": receipts})

sns.set(rc={'figure.figsize':(5,3.5)})

colors = ['#310783', '#078331']

sns_plot = sns.lineplot(data=data, x="users", y="receipts", color='#310783')
# suppress scientific notation
sns_plot.ticklabel_format(style='plain', axis='y')
sns_plot.set_title("Number of stored receipts by number of users")

sns_plot.figure.savefig("receipts_by_users_lineplot.pdf", bbox_inches = 'tight')
```
