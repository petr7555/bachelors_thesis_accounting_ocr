## Manual approach using means instead of passing the whole sentences and categories

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words="english")
# tokenize will:
#   - lowercase the sentence
#   - split it into words
#   - remove stop words
tokenize = vectorizer.build_analyzer()
```

```python
tokenize("I am so Happy today!")
```

```python
import numpy as np
import sys

def get_sentence_vector(vectors, sentence):
    """
    Returns n-dimensinal vector (n depends on the word embeddings model, usually 300).
    The vector is a mean of the vectors of words in the sentence.
    """
    tokens = tokenize(sentence)
    if len(tokens) == 0:
        sys.exit(f"Invalid sentence: {sentence}. It was either empty or composed solely of stop words.")
    mean_vec = np.mean(vectors.query(tokens), axis=0)
    return mean_vec
```

```python
get_sentence_vector(g_light, "I am so Happy today!").shape
```

```python
def get_category_using_mean_for_sentence_and_categories(vectors, sentence):
    """
    Computes mean vectors for all categories.
    Computes mean vector for the sentence.
    Computes similarities of the sentence mean vector to all categories' mean vectors.
    Picks the category with the largest similarity.
    """
    categories_vecs = [get_sentence_vector(vectors, key) for key in categories.keys()]
    mean_vec = get_sentence_vector(vectors, sentence)

    similarities = [vectors.similarity(mean_vec, category_vec) for category_vec in categories_vecs]
    most_similar_index = np.array(similarities).argmax()
    most_similar = list(categories)[most_similar_index]
    emoji = categories[most_similar]
    return {
        "category": most_similar,
        "emoji": emoji,
    }
```

## Compare 2 approaches:
  - when the whole sentences are vectorized
  - when they are split into words and the mean is taken

```python
def approach_comparison(items):
    data = {}

    for item in items:
        data[item] = []
        for approach in [get_whole_sentence_category, get_category_using_mean_for_sentence_and_categories]:
            result = approach(f_med, item)
            data[item].append(f"{result['emoji']} {result['category']}")

    df = pd.DataFrame.from_dict(data, 
                                columns=["get_whole_sentence_category", "get_category_using_mean_for_sentence_and_categories"],
                                orient='index')
    return df
```
